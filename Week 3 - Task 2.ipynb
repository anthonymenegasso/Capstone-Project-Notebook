# 1. Copying the work done before in Task 1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

```python
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3.head()
```

```python
df4 = df3[df3.Borough != 'Not assigned']
df4.head()
```

```python
df5 = df4.groupby(['Postal Code', 'Borough'], sort = False).agg(','.join)
df5.reset_index(inplace = True)
df5.head()
```

# 2. Using the csv file provided by Coursera

```python
url = "http://cocl.us/Geospatial_data"
df6 = pd.read_csv(url)
df6.head()
```

# 3. Combining the two Data Frames (df5 and df6) as latandlong

```python
latandlong = pd.merge(df5, df6, on='Postal Code')
latandlong.head(12)
```
